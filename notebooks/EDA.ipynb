```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
#loading library
cars = pd.read_csv('vehicles_us.csv')
cars.tail()
```

```python
#checking for duplicates
cars.duplicated().sum()
```

```python
#checking for missing data
cars.isnull().sum()
```

```python
#Graph idea cars based on type of fuel (diesel, other, gas, hybrid, electric)
```

```python
#looking at 4wd drive column 
cars['is_4wd'].unique()
```

We went ahead and checked what was in the 4wd column and we got two results. 1 and nan. So based on that we are going to assume nan is no 4wd and 1 is 4wd.

```python
#filling the missing paint colors

cars['paint_color'] = cars['paint_color'].fillna('unknown')
#Checking All THE Colors 
cars['paint_color'].unique()
```

```python
#making a graph color of the cars types

px.histogram(cars, x='type', color='paint_color',  title='How common colors are on vehicle types')
```

```python
#compare car models based on number of cylinders
px.histogram(cars, x='model',  color='cylinders',title='Cylinder Type on vehicle brands')
```

```python
px.scatter(cars, x='model_year', y='odometer', color='model', title='Odometer of vehicles')
```
